```python
from torchvision.datasets import MNIST
import torchvision.transforms as tfs
import pandas as pd
from torch.utils.data import DataLoader
import torch

```

```python
transforms = tfs.Compose([
    tfs.ToTensor()
])


root = './'
train = MNIST(root, train=True,  transform=transforms, download=True)
test  = MNIST(root, train=False, transform=transforms, download=True)
```

```python
batch_size = 128
train_loader = DataLoader(train, batch_size = batch_size, drop_last = True)
test_loader = DataLoader(test, batch_size = batch_size, drop_last = True)
```

```python
x_batch, y_batch = next(iter(train_loader))
```

```python
features = 28*28
classes = 10

w = torch.FloatTensor(features, classes).uniform_(-1, 1)
w.requires_grad_()
```

```python
epochs = 10
lr=0.3
```

```python
from torch import nn
criterion = nn.CrossEntropyLoss()
```

```python
print(x_batch.shape)
x_batch = x_batch.reshape(x_batch.shape[0], -1)
x_batch.shape
hist = []
```

```python
for i in range(epochs):
  for x_batch, y_batch in train_loader:
    
    # загружаем батч данных (вытянутый в линию)
    x_batch = x_batch.reshape(x_batch.shape[0], -1)

    # вычисляем функцию потерь (log loss, также известная как кросс-энтропия)
    logits = x_batch @ w
    probabilities = torch.exp(logits) / torch.exp(logits).sum(dim=1, keepdims=True)
    
    loss = -torch.log(probabilities[range(batch_size), y_batch]).mean()
    hist.append(loss.item())

    # вычисляем градиенты
    loss.backward()

    # делаем шаг градиентного спуска
    grad = w.grad
    with torch.no_grad():
      w -= lr * grad
    w.grad.zero_()

  print(f'{i+1},\t loss: {hist[-1]}')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 7))

plt.plot(hist)

plt.title('Loss by batch iterations')
plt.ylabel('Entropy Loss')
plt.xlabel('batches')

plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = 0
batches = 0

for x_batch, y_batch in test_loader:
    batches += 1
    x_batch = x_batch.view(x_batch.shape[0], -1)
    y_batch = y_batch

    preds = torch.argmax(x_batch @ w, dim=1)
    acc += (preds==y_batch).cpu().numpy().mean()

print(f'Test accuracy {acc / batches:.3}')
```
