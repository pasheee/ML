```python
def convmodel(X, W):
    horrange = 26
    vertrange = 26
    resultpic = torch.FloatTensor(vertrange, horrange).uniform_(-1, 1)
    for i in range(vertrange):
        for j in range(horrange):
            
            horind = j
            vertind = i
            whorind = 0
            wvertind = 0
            summ = 0.
            
            while vertind < i + 3:
                while horind < j + 3:
                    
                    #print(horind)
                    summ += int(W.data[wvertind][whorind] * X[0][vertind][horind])
                    
                    horind+=1
                    whorind+=1
                    
                vertind+=1
                wvertind+=1
                
            resultpic[i][j] = summ
    
    return resultpic
            
            
```
