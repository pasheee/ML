```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy.interpolate import make_interp_spline
from scipy.interpolate import interp1d
```

```python
x = [0, 0.05, 0.25, 1.0, 3.0]
y = [-10, -5, 10, 78, 227]
# y = [elem + 0.4 for elem in y]

if len(x) != len(y):
    raise ValueError('lens are not the same')
    
points = [(x[i], y[i]) for i in range(len(x))]
points = sorted(points, key = lambda x: x[0])
print(points)
x = np.array([points[i][0] for i in range(len(points))])
y = np.array([points[i][1] for i in range(len(points))])
```

```python
# Вычисление стандартной ошибки
standard_error = np.std(y) / np.sqrt(len(y))
# Приборная погрешность по оси х
inst_error = 0.1

# Значение t-критерия для доверительного интервала 95% и n-1 степеней свободы
t_value_y = stats.t.ppf(0.7, df=len(y) - 1)
t_value_x = stats.t.ppf(0.7, df = len(x) - 1)

# Погрешности
y_error = t_value_y * standard_error
x_error = t_value_x * inst_error 

# Интерполяция для гладкой кривой
x_new = np.linspace(x.min(), x.max(), 300)
# spl = scipy.interpolate.UnivariateSpline(x, y)
spl = interp1d(x, y, kind='quadratic')
y_smooth = spl(x_new)
```

```python
plot_name = 'Световая характеристика.'
x_label = 'H'
y_label = 'I'
```

```python
# Построение графика
plt.figure(figsize=(19, 10))
plt.plot(x_new, y_smooth, color = 'blue')
plt.errorbar(x, y, xerr = x_error, yerr=y_error, fmt='o', label='Погрешности по Стьюденту', color = 'blue')
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(plot_name)
plt.legend()
#Manually set ticks (Example: 10 ticks between min and max of x)
min_x = np.min(x)
max_x = np.max(x)
rounded_ticks = np.arange(min_x*100, max_x*100 + 10, 10)
plt.xticks(rounded_ticks/100) # Adjust 10 to your desired number of ticks

#Manually set ticks (Example: 10 ticks between min and max of y)
min_y = np.min(y)
max_y = np.max(y)

rounded_ticks = np.arange(min_y, max_y + 10, 10)

plt.yticks(rounded_ticks) 

plt.grid(which='both', axis='both', linestyle='--') 
plt.savefig("savedplots/lab5.2_lumen.png")
plt.show()
```
