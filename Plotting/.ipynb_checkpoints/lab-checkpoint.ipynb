#### Импорты

```python
import ROOT
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects
import matplotlib.ticker
import efc
```

#### Входные данные

```python
pd = [0.096, 0.108, 0.120, 0.135, 0.150, 0.168, 0.189, 0.213, 0.240, 0.267, 0.300, 0.330, 0.375, 0.420, 0.480, 0.540, 0.600, 0.660, 0.750, 0.840, 0.960]
pd = [el * 2 for el in pd]
U = [718, 692, 696, 681, 666, 648, 644, 630, 650, 663, 670, 723, 750, 769, 785, 803, 824, 835, 875, 903, 986]
U = [U[i] + 50 if i < 11 else U[i] for i in range(len(U))]
len(pd) - len(U)
print(U)
```

#### Расчёт погрешностей

```python
formula_for_pd = "p * d"
l = len(pd)
data_for_pd = {"p": [300] * l, "p_err": [0.1] * l,
              "d": [el / 300 for el in pd], "d_err": [0.01 / 1000] * l} 
dpd, formula_for_pd = efc.get_errors(formula_for_pd, data_for_pd)

dU = [1] * l
```

#### Данные для нанесения на график

```python
x = pd.copy() # значения, которые пойдут на график
y = U.copy()
dx = dpd.copy()
dy = dU.copy()
```

#### Фитирование с помощью ROOT

```python
def formula(x, par):
    # x = x[0]
    # a = par[0]
    # mu = par[1]
    # sigma = par[2]
    x = x[0]
    A = par[0]
    B = par[1]
    gamma = par[2]

    return B * x / np.log(A * x / np.log(1 + 1 / gamma))


    return a * np.exp(-(x - mu) ** 2 / sigma ** 2)


c = ROOT.TCanvas()

graph = ROOT.TGraphErrors() # точки с погрешностями

for i in range(len(x)):
    graph.SetPoint(i, x[i], y[i])
    graph.SetPointError(i, dx[i], dy[i])

fit = ROOT.TF1("fit", "pol6", x[0], x[-1])
fit.SetParameter(0, 20)
fit.SetParameter(1, 487)
fit.SetParameter(2, 0.05) # sigma
graph.Draw("A*")
graph.Fit(fit, "S")
```

```python
# c.Update()
c.Draw()
```

```python
def fit_function(x, par): # фитирующая функция с параметрами от рута
    return sum(x ** i * par[i] for i in range(len(par)))
par_number = 7
par = [fit.GetParameter(i) for i in range(par_number)]
x_fit = np.linspace(min(x) - 1.1 * max(dx), max(x) + 1.1 * max(dx), 100) # точки для нанесения на график
y_fit = fit_function(x_fit, par)

def theory(x, par): # фитирующая функция с параметрами от рута
    A = par[0]
    B = par[1]
    gamma = par[2]
    return B * (x + 0.4) / np.log(A * (x + 0.4) / np.log(1 + 1 / gamma)) - 35
par = [20, 700, 0.00051]
y_th_fit = theory(x_fit, par)
```

#### МНК

```python
def mse(x, y):
    xy = [x[i] * y[i] for i in range(len(x))]
    x2 = [x[i] ** 2 for i in range(len(x))]
    
    k = (len(x) * sum(xy) - sum(x) * sum(y)) / (len(x) * sum(x2) - sum(x) ** 2)
    b = (sum(y) - k * sum(x)) / len(x)
    
    if b > 0:
        mu = -1 / ((k ** 2 + 1) ** 0.5)
    else:
        mu = 1 / ((k ** 2 + 1) ** 0.5)
    
    sum_delta = 0
    for i in range(len(x)):
        sum_delta += (abs(mu * (k * x[i] - y[i] + b)))

    dk = np.sqrt(1 / (len(x) - 2) * ((np.mean([el ** 2 for el in y]) - np.mean(y) ** 2) / (np.mean([el ** 2 for el in x]) - np.mean(x) ** 2) - k ** 2))
    db = dk * np.sqrt(np.mean([el ** 2 for el in x]))
    
    return [k, b, dk, db]

k, b, dk, db = mse(x, y)
print(f"k = {k} +/- {dk} \nb = {b} +/- {db}")
```
