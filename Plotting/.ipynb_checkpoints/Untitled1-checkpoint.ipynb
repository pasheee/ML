```python
def theory_f(x, gamma, A=112.5, B=2737.5):
    f =  (B * x) / (np.log(A * x) - np.log(np.log(1 + 1/gamma)))
    return f
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np
```

```python
# pd = [0.096, 0.108, 0.120, 0.135, 0.150, 0.168, 0.189, 0.213, 0.240, 0.267, 0.300, 0.330, 0.375, 0.420, 0.480, 0.540, 0.600, 0.660, 0.750, 0.840, 0.960]
# U = [642, 570, 554, 543, 505, 479, 454, 456, 478, 499, 504, 520, 530, 540, 554, 576, 586, 603, 625, 645, 667]
pd = [0.24, 0.27, 0.3, 0.336, 0.378, 0.426, 0.48, 0.534, 0.6, 0.66, 0.75, 0.84, 0.96, 1.08, 1.2, 1.32, 1.5]
U =  [538, 508, 492, 473, 490, 502, 506, 516, 523, 527, 540, 558, 579, 597, 613, 627, 650]
points = []
for i in range( len(pd) ):
    points.append((pd[i], U[i]))
    
theory_pd = np.array([0.096, 0.108, 0.120, 0.135, 0.150, 0.168, 0.189, 0.213, 0.24, 0.27, 0.3, 0.336, 0.378, 0.426, 0.48, 0.534, 0.6, 0.66, 0.75, 0.84, 0.96, 1.08, 1.2, 1.32, 1.5])

points = sorted(points)

pd = np.array( [x for x, _ in points] )
U = np.array( [y for _, y in points] )
```

```python
f_ur = interp1d(pd, U, kind='quadratic')
pd_sample = np.linspace(pd.min(), pd.max(), 100)
pd_sample_theory = np.linspace(theory_pd.min(), theory_pd.max(), 100)
theory_u = np.array([theory_f(x, 0.0008) for x in pd_sample_theory])
U_sample = f_ur(pd_sample)
```

```python
plt.figure(figsize = (19, 10))
plt.locator_params(axis='x', nbins=30)
plt.locator_params(axis='y', nbins=50)

plt.errorbar(pd_sample, U_sample, label = 'U')
plt.errorbar(pd_sample, theory_u, label = 'Theory')
plt.scatter(pd, U)


plt.title('             6')
plt.ylabel('U, B')
plt.xlabel('pd,  * ')
plt.grid()
plt.legend()
plt.savefig("savedplots/lab5_16_6gP.png")
plt.show()
```
