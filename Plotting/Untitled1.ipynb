```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np
```

```python
x = [4000, 1560, 5600, 14500, 24000, 38000, 49400, 56700, 43500, 32300, 28200, 6400, 2900]
Ur = [1.09, 1.72, 1.00, 0.96, 1.00, 1.14, 1.35, 1.46, 1.32, 1.11, 1.06, 0.95, 1.20]
phi = [0, 22, 29, 17, 8.6, 5.2, -6, -16, -26, -30, -31, -28, -23, -21, 15, 26]


pr = []
pphi = []
for i in range(len(x)):
    pr.append((x[i], Ur[i]))
    pphi.append((x[i], phi[i]))

pr = sorted(pr)
pphi = sorted(pphi)

freq = np.array([x for x, _ in pr])
ur = np.array([y for _, y in pr])
phi = [y for _, y in pphi]
```

```python
f_ur = interp1d(freq, ur, kind='quadratic')
freq = np.linspace(freq.min(), freq.max(), 500) # 500 points for smooth curve

ur = f_ur(freq)
```

```python
plt.figure(figsize = (19, 10))
plt.errorbar(freq, ur, label = 'UR')
# plt.errorbar(freq, ul, label = 'UL')
# plt.errorbar(freq, uc, label = 'Uc')
plt.title('График зависимости Ur от частоты в параллельном контуре')
plt.ylabel('U, B')
plt.xlabel('frequency, Hz')
plt.grid()
# plt.legend()
plt.savefig("savedplots/eltechlab2_forth.png")
plt.show()
```
