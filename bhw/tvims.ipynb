```python
# TASK 3

# Unbiased estimation of mathematical expectation (average value)
mean = np.mean(data)

# Unbiased variance estimation
variance = np.var(data, ddof=1)

print("Mathematical expectation:", mean)
print("Variance:", variance)
```

```python
# TASK 4

# Параметры нормального распределения
mean_moment = mean
variance_moment = variance

print("Estimation of mathematical expectation by the method of moments:", mean_moment)
print("Estimation of variance by the method of moments:", variance_moment)
```

```python
def negative_log_likelihood(params, data):
    """
    A function for calculating the negative logarithm of the likelihood function.
    params[0] - mathematical expectation (μ)
    params[1] - standard deviation (σ)
    """
    mu, sigma = params
    n = len(data)

    if sigma <= 0:
        return np.inf

    log_likelihood = -n * np.log(1/(2 * np.pi) ** 0.5) - n * np.log(1/sigma) + \
                     np.sum((data - mu)**2) / (2 * sigma**2)

    return log_likelihood 
def maximum_likelihood_estimation(data):
    initial_guess = [0, 0]

    result = minimize(negative_log_likelihood, 
                     initial_guess, 
                     args=(data,),
                     method='Nelder-Mead')

    return result.x

mu_mle, sigma_mle = maximum_likelihood_estimation(data)

var_mle = sigma_mle**2
    

print("Estimation of mathematical expectation by the maximum likelihood method:", mu_mle)
print("Maximum likelihood estimation of variance:", var_mle)
```

```python
def confidence_intervals(data, confidence=0.9):
    n = len(data)
    mean = np.mean(data)
    var = np.var(data, ddof=1)
    
    # Для мат. ожидания
    t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
    mean_margin = t_value * np.sqrt(var / n)
    mean_interval = (mean - mean_margin, mean + mean_margin)
    
    # Для дисперсии
    chi2_lower = stats.chi2.ppf((1 - confidence) / 2, df=n-1)
    chi2_upper = stats.chi2.ppf((1 + confidence) / 2, df=n-1)
    var_interval = ((n-1) * var / chi2_upper, (n-1) * var / chi2_lower)

    return mean_interval, var_interval

print("6. Confidence intervals (90%)")
mean_interval, var_interval = confidence_intervals(data)
print(f"Интервал для мат. ожидания: ({mean_interval[0]:.2f}, {mean_interval[1]:.2f})")
print(f"Интервал для дисперсии: ({var_interval[0]:.2f}, {var_interval[1]:.2f})\n")
```

```python
import numpy as np
from scipy.stats import norm, chisquare

array = [
    0.11, 2.83, -0.26, -1.2, -3.36, 1.5, -2.12, 2.68, -4.16, -2.43,
    -1.58, -1.72, -3.69, 0.19, -0.55, -1.46, 0.68, 2.28, -4.09, 1.71,
    3.39, -0.68, -0.4, -3.86, 1.3, -2.28, 0.41, 2.22, 1.26, -1.62,
    1.88, 0.21, -1.96, -3.63, 1.41, 1.67, -2.19, -1.37, 0.25, -0.3,
    -2.8, -5.86, -1.32, -3.99, 1.27, 1.94, -2.19, -1.82, -0.05, 0.14
]


data = np.array(array)


num_bins = 6


hist, bin_edges = np.histogram(data, bins=num_bins)


mean = np.mean(data)
variance = np.var(data, ddof=1)
std_dev = np.sqrt(variance)


probabilities = np.array([norm.cdf(bin_edges[i+1], mean, std_dev) - 
                         norm.cdf(bin_edges[i], mean, std_dev) 
                         for i in range(num_bins)])


total_count = len(data)
expected_freq = probabilities * total_count


expected_freq = expected_freq * (np.sum(hist) / np.sum(expected_freq))

print("The sum of the observed frequencies:", np.sum(hist))
print("The sum of the expected frequencies:", np.sum(expected_freq))


chi2_stat, p_value = chisquare(hist, expected_freq)


df = num_bins - 3  

alpha = 0.05

print("\nTest results:")
print(f"Chi-square statistics: {chi2_stat:.4f}")
print(f"Degrees of freedom: {df}")

if p_value < alpha:
    print("\nWe reject the null hypothesis: the data is not distributed normally.")
else:
    print("\nWe do not reject the null hypothesis: the data is distributed normally.")

print("\nDetailed information:")
print("Observed frequencies:", hist)
print("Expected frequencies:", expected_freq)
```
